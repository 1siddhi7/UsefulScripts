```
#loading the zipped folder of dataset uploaded in My Drive
dataset_archive = zipfile.ZipFile('/content/drive/My Drive/data_traffic_images.zip', 'r')

#reading and displaying a single image from the zipped archive without extracting the entire folder
img_data = dataset_archive.read('data/frame0.jpg')
display(Image(img_data))
```

```
#loading all images from the zipped archive in a numpy array without extracting the entire folder

from PIL import Image

data_array = []

for i in range(209):
    img_data = dataset_archive.read('data/frame'+str(i)+'.jpg')
    image = np.array(Image.open(io.BytesIO(img_data))) 
    data_array.append(image)

data_array = np.array(data_array)
print('shape of data_array: ',data_array.shape)
```
